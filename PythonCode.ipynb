```python
import matplotlib.pyplot as plt
import pandas as pd

# Enrollment data
data = {
    'District': ['Nankana Sahib', 'Bahawalpur', 'Rahim Yar Khan', 'Jhang', 'Attock', 'Sahiwal'],
    'Group': ['Treated', 'Treated', 'Treated', 'Control', 'Control', 'Control'],
    'Enrollment_2010': [212077, 285800, 497166, 306434, 217637, 300392],
    'Enrollment_2011': [169895, 253470, 469256, 302352, 216825, 299256]
}

df = pd.DataFrame(data)

# Plot enrollment change for each district
fig, ax = plt.subplots(figsize=(10, 6))
x = range(len(df))

ax.bar(x, df['Enrollment_2010'], width=0.4, label='2010 Enrollment', align='center')
ax.bar([p + 0.4 for p in x], df['Enrollment_2011'], width=0.4, label='2011 Enrollment', align='center')

ax.set_xticks([p + 0.2 for p in x])
ax.set_xticklabels(df['District'], rotation=45, ha='right')
ax.set_ylabel('Enrollment')
ax.set_title('Enrollment by District (2010 vs 2011)')
ax.legend()

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```
